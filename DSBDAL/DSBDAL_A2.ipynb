```python
label_encoder = LabelEncoder()
df1['Gender_encoded'] = label_encoder.fit_transform(df1['gender'])
```

```python
temp = pd.get_dummies(df1["lunch"])
```

```python
df1 = pd.concat([df1, temp['standard']], axis='columns')
```

```python
df1.drop(['lunch'], axis='columns', inplace=True)
```

```python
df1.shape
```

```python
Q1 = df1["math score"].quantile(0.25)
Q3 = df1["math score"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

upper_array = np.where(df1['math score'] >= upper_limit)[0]
lower_array = np.where(df1['math score'] <= lower_limit)[0]

df1.drop(index=upper_array, inplace=True)
df1.drop(index=lower_array, inplace=True)
```

```python
df1.shape
```

```python
Q1 = df1["reading score"].quantile(0.25)
Q3 = df1["reading score"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

upper_array = np.where(df1['reading score'] >= upper_limit)[0]
lower_array = np.where(df1['reading score'] <= lower_limit)[0]

df1.drop(index=upper_array, inplace=True)
df1.drop(index=lower_array, inplace=True)
```

```python
df1.shape
```

```python
Q1 = df1["writing score"].quantile(0.25)
Q3 = df1["writing score"].quantile(0.75)

IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

upper_array = np.where(df1['writing score'] >= upper_limit)[0]
lower_array = np.where(df1['writing score'] <= lower_limit)[0]

df1.drop(index=upper_array, inplace=True)
df1.drop(index=lower_array, inplace=True)
```
